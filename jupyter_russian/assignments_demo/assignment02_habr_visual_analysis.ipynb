Избавимся сразу от переменных, названия которых заканчиваются на `_lognorm` (нужны для соревнования на Kaggle). Выберем их с помощью `filter()` и удалим `drop`-ом:

```python
df.drop(filter(lambda c: c.endswith('_lognorm'), df.columns), 
        axis = 1,       # axis = 1: столбцы 
        inplace = True) # избавляет от необходимости сохранять датасет
```

```python
df.describe(include = ['object', 'bool'] # бинарные и категориальные переменные
           ).T
```

```python
# настройка внешнего вида графиков в seaborn
sns.set_style("dark")
sns.set_palette("RdBu")
sns.set_context("notebook", font_scale = 1.5, 
                rc = { "figure.figsize" : (15, 5), "axes.titlesize" : 18 })
```

Столбец **`published`** (время публикации) содержит строки. Чтобы мы могли работать с этими данными как с датой/временем публикации, приведём их к типу `datetime`:

```python
print(df.published.dtype)
df['published'] = pd.to_datetime(df.published, yearfirst = True)
print(df.published.dtype)
```

Создадим несколько столбцов на основе данных о времени публикации:

```python
df['year'] = [d.year for d in df.published]
df['month'] = [d.month for d in df.published]

df['dayofweek'] = [d.isoweekday() for d in df.published]
df['hour'] = [d.hour for d in df.published]
```

## 3\. Когда лучше всего публиковать статью?

* Больше всего просмотров набирают статьи, опубликованные в 12 часов дня
* У опубликованных в 10 утра постов больше всего комментариев
* Больше всего просмотров набирают статьи, опубликованные в 6 часов утра

```python
data = df.groupby('hour')['views'].mean()
data.plot(kind='bar', rot=75, figsize=(12,8))
plt.title('Диаграмма среднего количества просмотров по часам')
plt.xlabel('Часы')
plt.ylabel('Среднее количество просмотров')
plt.show()
```

```python
data = df.groupby('hour')['comments'].mean()
data.plot(kind='bar', rot=75, figsize=(12,8))
plt.title('Диаграмма среднего количества комментариев к статье')
plt.xlabel('Часы')
plt.ylabel('Среднее количество комментариев')
plt.show()
```

```python
data = df[df['domain']=='geektimes.ru']
data = data.sort_values(by='comments', ascending=False)[:100]
plt.figure(figsize=(12,8))
plt.scatter(data.hour,data.comments)
plt.title('Диаграмма количества комментариев\n \
к топ-100 статьям на geektimes.ru')
plt.xlabel('Часы')
plt.ylabel('Количество комментариев')
plt.show()
```

### Ответ:

## 4\. Кого из топ-20 авторов чаще всего минусуют?

* @Mordatyj

```python
df['votes']=df.votes_plus - df.votes_minus
data = df.groupby('author')[['votes']].sum()
data = data.sort_values(by='votes', ascending=False)[:20]
dt = df.loc[df['author'].isin(data.index)]
dt = dt.groupby('author')['votes_minus'].sum()
dt.plot(kind='bar', rot=75, figsize=(12,8))
plt.title('Количество дизлайков по авторам из ТОП-20')
plt.xlabel('Авторы из ТОП-20')
plt.ylabel('Количество дизлайков')
plt.show()
```

## 5\. Сравните субботы и понедельники

Правда ли, что по субботам авторы пишут в основном днём, а по понедельникам — в основном вечером?

```python
df.head()
data = df.loc[df['dayofweek'].isin([1,6])]
_, axes = plt.subplots(figsize=(12,8))
sns.countplot(
    ax=axes,
    x=data.hour, 
    hue='dayofweek', 
    data=data
)
plt.show()
```

### Ответ:
Диаграмма распределения количества статей по часам в понедельник имеет явно выраженную моду в 12 часов дня. В субботу распределение имеет моды в 15 часов и 22 часа.
